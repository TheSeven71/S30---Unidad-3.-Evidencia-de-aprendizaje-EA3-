# Paso 1. Diseño del esquema de almacenamiento

## Descripción del conjunto de datos

El conjunto de datos contiene información sobre el uso de herramientas de Inteligencia Artificial por estudiantes universitarios, incluyendo variables académicas, hábitos de estudio, nivel de dependencia de IA, rendimiento académico y riesgo de agotamiento (burnout).

La entidad principal es **Estudiante**.

````python
from pyspark.sql.types import *

schema = StructType([
    StructField("Student_ID", LongType(), False),
    StructField("Major_Category", StringType(), False),
    StructField("Year_of_Study", StringType(), False),
    StructField("Pre_Semester_GPA", DoubleType(), False),
    StructField("Weekly_GenAI_Hours", DoubleType(), False),
    StructField("Primary_Use_Case", StringType(), False),
    StructField("Prompt_Engineering_Skill", StringType(), False),
    StructField("Tool_Diversity", IntegerType(), False),
    StructField("Paid_Subscription", BooleanType(), False),
    StructField("Traditional_Study_Hours", DoubleType(), False),
    StructField("Perceived_AI_Dependency", IntegerType(), False),
    StructField("Institutional_Policy", StringType(), False),
    StructField("Anxiety_Level_During_Exams", IntegerType(), False),
    StructField("Post_Semester_GPA", DoubleType(), False),
    StructField("Skill_Retention_Score", DoubleType(), False),
    StructField("Burnout_Risk_Level", StringType(), False)
])

print(schema)
````

````python
%sql
DESCRIBE TABLE workspace.default.ai_student;
````

Esta consulta permite verificar la estructura de la tabla, mostrando el nombre de cada columna, su tipo de dato y otra información relevante. Su propósito es validar que el esquema cargado coincide con el diseño definido previamente.

````python
%sql
SHOW CREATE TABLE workspace.default.ai_student;
````

Permite visualizar la sentencia de creación de la tabla. Se utiliza para validar cómo Databricks almacenó la estructura y configuración del conjunto de datos.

````python
df = spark.table("workspace.default.ai_student")

df.printSchema()
````

Muestra el esquema de la tabla desde Spark, incluyendo los tipos de datos de cada atributo. Permite verificar la correcta interpretación de los datos durante la carga.

````python
df.describe().show()
````

Genera estadísticas descriptivas básicas como cantidad de registros, promedio, desviación estándar, valor mínimo y máximo para las variables numéricas. Esto permite detectar valores anómalos y validar la calidad de los datos.

````python
%sql

SELECT
COUNT(*) AS total_registros,
AVG(Pre_Semester_GPA) AS promedio_gpa_inicial,
AVG(Post_Semester_GPA) AS promedio_gpa_final,
AVG(Weekly_GenAI_Hours) AS promedio_horas_ia
FROM workspace.default.ai_student;
````

Se calculan métricas generales del dataset para validar la consistencia de la información y obtener una visión inicial de los datos almacenados.

````python
%sql

SELECT *
FROM workspace.default.ai_student
LIMIT 10;
````

Permite inspeccionar manualmente una muestra de registros para verificar que los datos fueron cargados correctamente.

````python
df.show(10)
````

Equivalente a la consulta SQL anterior, mostrando los primeros registros mediante la API de Spark.

````python
%sql

SELECT
Major_Category,
COUNT(*) AS cantidad_estudiantes
FROM workspace.default.ai_student
GROUP BY Major_Category
ORDER BY cantidad_estudiantes DESC;
````

Permite conocer cuántos estudiantes pertenecen a cada área académica y validar que los datos categóricos fueron cargados correctamente.

````python
df.groupBy("Major_Category") \
  .count() \
  .orderBy("count", ascending=False) \
  .show()
````

Realiza el mismo análisis utilizando Spark DataFrame, permitiendo comparar resultados con SQL.

````python
%sql

SELECT COUNT(*) AS total_estudiantes
FROM workspace.default.ai_student;
````

Verifica el número total de registros almacenados en la tabla.

````python
df.count()
````

Equivalente al COUNT(*) de SQL para validar la cantidad de registros cargados.

````python
%sql

SELECT *
FROM workspace.default.ai_student
WHERE Weekly_GenAI_Hours > 20
LIMIT 10;
````

Permite validar consultas con condiciones específicas y analizar subconjuntos de datos.

````python
df.filter(df.Weekly_GenAI_Hours > 20).show(10)
````

Equivalente al filtro SQL utilizando la API de Spark.

Las validaciones realizadas mediante Spark SQL y PySpark permitieron confirmar la correcta carga del conjunto de datos, la consistencia del esquema definido y la disponibilidad de los registros para análisis posteriores. Los resultados obtenidos mediante consultas SQL y operaciones DataFrame fueron equivalentes, demostrando que ambas interfaces permiten acceder y procesar la misma información almacenada en Databricks.
